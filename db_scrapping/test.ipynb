```python
import requests
from bs4 import BeautifulSoup

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
    
}


html_tables = []

for zone in  time_zones:

    for year in range(21,101): 

        url = base_link + str(20 * ((year-1) // 20 ) + 1) + "/almanac" + str(2000 + year) + zone.lower() + ".html"

        # Send a GET request to the webpage
        response = requests.get(url, headers = headers)


        # Check if the request was successful
        if response.status_code == 200:
            # Get the HTML content of the webpage
            html_content = response.text

            # Create a BeautifulSoup object and specify the parser
            soup = BeautifulSoup(html_content, 'lxml')

            data_tab = soup.find('table', class_ = "datatab")
            html_tables.append(data_tab)
            
        else:
            print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
            print(url)
        


```

```python
def event_extractor(data_table, time_zone = "UTC"):

    # Add only event elements 
    events_html = data_table.find_all("pre")
    


    pre_text = events_html[0].get_text() + events_html[1].get_text() 


    months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    # Process the text line by line
    lines = pre_text.split('\n')
    
    events = []
    current_month = None
    current_date = None

    for line in lines:
        # Skip empty lines
        if not line.strip():
            continue

        # Check if the line starts with a month abbreviation
        if line[:3] in months:
            parts = line.split(maxsplit=3)
            current_month = parts[0]
            current_date = f"{current_month} {parts[1]}"
            time_event = parts[2] if len(parts) > 2 else "00:00"
            event = parts[3]
            events.append((current_date, time_event, event))
        elif current_month:
            # Line contains time and event details
            parts = line.split(maxsplit=3)
            current_date = parts[0] if len(parts) > 1 else ""
            time_event = parts[1] if len(parts) > 1 else "00:00"
            try:
                event = parts[2]
            except:
                print(line)
            events.append((current_month + current_date, time_event,event))

    return events
         
```
